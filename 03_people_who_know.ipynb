```python
import json
def load_data(filename):
    with open(filename,"r") as f :
        return json.load(f)
def find_people_who_know(user_id,data):
    user_friends={}
    for user in data['users']:
         user_friends[user['id']]=set(user['friends'])

    if user_id not in user_friends:
        return []

    direct_friends=user_friends[user_id]
    suggetion={}
    for friend in  direct_friends:
        for mutual in  user_friends[friend]:
            if mutual!=user_id and mutual not in direct_friends:
                suggetion[mutual]=suggetion.get(mutual,0)+1
                
    sorted_suggetion=sorted(suggetion.items(),key=lambda x:x[1], reverse=True)
    return [user_id for user_id,mutual_count in  sorted_suggetion]
        
        


data=load_data("data2.json")
user_id=1
rec=find_people_who_know(user_id,data)
print(rec)
```
